```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('startup_funding.csv',encoding='utf-8',skipinitialspace=True)
```

```python
data
```

```python
data['year']=[int(float(i[-4:])) for i in data['Date']]
data
```

```python
#yearwise funding
yr=data['year'].value_counts()
print(yr)
yr.sort_index(inplace=True)
print(yr)
```

```python
#Top Indian Cities
data['CityLocation']=data['CityLocation'].replace('bangalore','Bangalore')
data['CityLocation']=data['CityLocation'].replace('Delhi','New Delhi')

def correct(string):
    string=str(string)
    return string.split('/')[0].strip()
data['CityLocation']=data['CityLocation'].apply(correct)

array=data['CityLocation'].value_counts()
x=array.index
y=array.values
a=[]
b=[]
for i in range(11):
    if x[i]!='nan':
        print(x[i],y[i])
        a.append(x[i])
        b.append(y[i])

plt.pie(b,labels=a)
plt.show()
```

```python
#Funding amount

data.dropna(inplace=True,subset=['AmountInUSD'])
data.dropna(inplace=True,subset=['CityLocation'])

def correct(string):
    string=str(string)
    return string.split('/')[0].strip()
data['CityLocation']=data['CityLocation'].apply(correct)
data['CityLocation']=data['CityLocation'].replace('bangalore','Bangalore')
data['CityLocation']=data['CityLocation'].replace('Delhi','New Delhi')


data['AmountInUSD']=data['AmountInUSD'].apply(lambda x: float(str(x).replace(',','')))


c = data.groupby("CityLocation")['AmountInUSD'].sum().sort_values(ascending = False)[:10]
d = c.index
e = c.values

a=[]
b=[]

df={}
for i in range(len(c)):
    df[d[i]]=(e[i]/sum(e))*100
    a.append(d[i])
    b.append(e[i])
for i,j in df.items():
    print(i,'{:.2f}'.format(j))

plt.pie(b, labels = a, autopct = "%.2f%%")
plt.show()
```

```python
#Investment Type
data.dropna(inplace = True,subset=['AmountInUSD'])
data["AmountInUSD"] = data["AmountInUSD"].apply(lambda x: float(str(x).replace(",","")))

data.dropna(inplace = True,subset=['InvestmentType'])
data['InvestmentType'].replace("Crowd funding", 'Crowd Funding', inplace = True)
data['InvestmentType'].replace('PrivateEquity', 'Private Equity', inplace = True)
data['InvestmentType'].replace('SeedFunding', 'Seed Funding', inplace = True)


c = data.groupby("InvestmentType")['AmountInUSD'].sum().sort_values(ascending = False)
d = c.index
e = c.values

df={}
for i in range(len(d)):
    df[d[i]]=(e[i]/sum(e))*100
for i,j in df.items():
    print(i,'{:.2f}'.format(j))

plt.pie(e, labels = d, autopct = "%.2f%%")
plt.show()
```

```python
#Top Industries
d=dict()
for i in data['IndustryVertical']:
    d[i]=d.get(i,0)+1
    
data.dropna(inplace = True,subset=['AmountInUSD'])
data["AmountInUSD"] = data["AmountInUSD"].apply(lambda x: float(str(x).replace(",","")))

data.dropna(inplace=True,subset=['IndustryVertical'])
data["IndustryVertical"].replace('eCommerce','Ecommerce', inplace = True)
data['IndustryVertical'].replace("ECommerce","Ecommerce", inplace = True)
data['IndustryVertical'].replace("ecommerce","Ecommerce", inplace = True)

c = data.groupby("IndustryVertical")['AmountInUSD'].sum().sort_values(ascending = False)[:5]
d = c.index
e = c.values

for i in range(len(d)):
    k=((e[i]/sum(e))*100)
    print(d[i],'{:.2f}'.format(k))
```

```python
#Top Startups

data['StartupName'].replace('Olacabs','Ola',inplace=True) 
data['StartupName'].replace('Ola Cabs','Ola',inplace=True) 
data['StartupName'].replace('Flipkart.com','Flipkart',inplace=True) 
data['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True) 
data['StartupName'].replace('Oyo Rooms','Oyo',inplace=True) 
data['StartupName'].replace('Oyorooms','Oyo',inplace=True) 
data['StartupName'].replace('OyoRooms','Oyo',inplace=True) 
data['StartupName'].replace('OYO Rooms','Oyo',inplace=True) 

data.dropna(inplace = True,subset=['AmountInUSD'])
data["AmountInUSD"] = data["AmountInUSD"].apply(lambda x: float(str(x).replace(",","")))

c = data.groupby("StartupName")['AmountInUSD'].sum().sort_values(ascending = False)[:5]
d = c.index

for i in range(5):
    print(d[i])
```

```python
#funding rounds

data['StartupName'].replace('Olacabs','Ola',inplace=True) 
data['StartupName'].replace('Ola Cabs','Ola',inplace=True) 
data['StartupName'].replace('Flipkart.com','Flipkart',inplace=True) 
data['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True) 
data['StartupName'].replace('Oyo Rooms','Oyo',inplace=True) 
data['StartupName'].replace('Oyorooms','Oyo',inplace=True) 
data['StartupName'].replace('OyoRooms','Oyo',inplace=True) 
data['StartupName'].replace('OYO Rooms','Oyo',inplace=True) 

data['count']=1
c = data.groupby("StartupName")['count'].sum().sort_values(ascending = False)[:5]
d = c.index
e = c.values

for i in range(len(d)):
    print(d[i],e[i])
```
